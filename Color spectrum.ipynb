```python
points = [41,74,107,140]
x1 = range(points[0], points[1]+1)
x2 = range(points[1], points[2]+1)
x3 = range(points[2], points[3]+1)

# red
r1 = [200 - int((200.0/33)*(x - points[0])) for x in x1]
r2 = [int((255.0/33)*(x - points[2])) for x in x3]

# green
g1 = [int((255.0/33)*(x - points[1])) for x in x2]
g2 = [255 - int((255.0/33)*(x - points[2])) for x in x3]

# blue
b1 = [55 + int((200.0/33)*(x - points[0])) for x in x1]
b2 = [255 - int((255.0/33)*(x - points[1])) for x in x2]
```

```python
import matplotlib.pyplot as plt
plt.plot(x1,r1,'r',x3,r2,'r') # red
plt.plot(x2,g1,'g',x3,g2,'g') # green
plt.plot(x1,b1,'b',x2,b2,'b') # blue
plt.show()
```

```python
from PIL import Image
im = Image.new('RGB', (100, 50))
ld = im.load()

for x in x1:
    pos = x - points[0]
    i = x - points[0]
    for y in range(im.size[1]):
        ld[pos, y] = (r1[i], 0, b1[i])

for x in x2:
    pos = x - points[0]
    i = x - points[1]
    for y in range(im.size[1]):
        ld[pos, y] = (0, g1[i], b2[i])

for x in x3:
    pos = x - points[0]
    i = x - points[2]
    for y in range(im.size[1]):
        ld[pos, y] = (r2[i], g2[i], 0)

im.save('gradient.png')
```
